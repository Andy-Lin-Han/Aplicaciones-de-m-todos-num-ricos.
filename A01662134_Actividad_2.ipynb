Ejercicio 1:

Euler a "mano " con LateX


$\frac{dT}{dt} = -k(T-T_a) $, donde

$ k = 0.019 $

$ T(0) = 70 °C $

$T_a = 20 $ es decir

$\frac{dT}{dt} = -0.019(T-20) $

$ t \in [0,10] $ con $ h = 2 $  

Procedimiento :

Sea:


$ T(t + h) \approx T'(t) h + T(t) $

Con la condición inicial obtenemos lo siguiente:

$ T(0) = 70 °C$

$T(2) ≈ -0.019(T(0)-20) * 2 + T(0) ≈  68.1 °C $

$T(4) ≈ -0.019(T(2)-20) * 2 + T(2) ≈  66.272 °C $

$T(6) ≈ -0.019(T(4)-20) * 2 + T(4) ≈  64.514 °C $

$T(8) ≈ -0.019(T(6)-20) * 2 + T(6) ≈  62.822 °C $

$T(10) ≈ -0.019(T(8)-20) * 2 + T(8) ≈  61.195 °C $



Euler con código:

```python
h = 2

a =  0

b = 10

T_i = 70

tiempo = [0]

temperatura = [70]

print(" Cuando t=", 0,"min.", "T =" ,T_i, "°C")

for i in range (a+2 , b+1,2):

  T_prima = -0.019*(T_i -20)

  T_mas_h = T_i + T_prima*h

  print(" Cuando t=", i,"min.","T =" ,T_mas_h, "°C")

  T_i = T_mas_h

  tiempo.append(i)

  temperatura.append(T_mas_h)





```

Euler con gráfica:

```python
import numpy as np

import matplotlib.pyplot as graficar

eje_x_1 = np.array(tiempo)

eje_y_1 = np.array(temperatura)

graficar.figure(figsize=(8,8))

graficar.plot(eje_x_1, eje_y_1, label = "Sol. Aproximada", linewidth = 2)

graficar.title (" Solución aproximada con Euler")

graficar.xlabel("Tiempo (min)")

graficar.ylabel("Temperatura (°C)")

graficar.legend()

graficar.grid(True)

graficar.show()

```

$\frac{dT}{dt} = -k(T-T_a) $, donde

$ t \in [0,5] $ con  $Δt = 0.5 $  

Esto implica que, cuando:

$ T(0.5) ≈  -0.12(T(0)-10)*0.5+T(0) \approx 35.38 $

Euler con código.

```python
h = 0.5

a =  0

b = 5

T_i = 37

N = (b-a)/h

N = int(N)

tiempo_1 = [0]

temperatura_1 = [37]

print(" Cuando t=", 0,"hrs.", "T =" ,T_i, "°C")

for i in range (a+1 , N+1):

  T_prima = -0.12*(T_i -10)

  T_mas_h = T_i + T_prima*h

  print(" Cuando t=", i/2,"hrs.","T =" ,T_mas_h, "°C")

  tiempo_1.append(i/2)

  temperatura_1.append(T_mas_h)

  T_i = T_mas_h
```

Euler a "mano " con LateX

$\frac{dT}{dt} = -k(T-T_a) $, donde

$T_\alpha = -2t+20  $ es decir

$ t \in [0,5] $ con $ Δt = 0.5 $  

Esto implica que, cuando:

$ T(0) = 37 °C $

$ T(0.5) ≈  -0.12(T(0)-20+2(0))*0.5+T(0) \approx 35.98 $

$ T(1) ≈  -0.12(T(0.5)-20+2(0.5))*0.5+T(0.5) \approx 34.9612 $

Euler con código:

```python
h = 0.5

a =  0

b = 5

T_i = 37

N = (b-a)/h

N = int(N)

tiempo_2 = [0]

temperatura_2 = [37]

print(" Cuando t=", 0,"hrs.", "T =" ,T_i, "°C")

for i in range (a , N):

  tempo = [0,0.5,1,1.5,2,2.5,3,3.5,4,4.5]

  T_a = 20-2*tempo[i]

  T_prima = -0.12*(T_i -T_a)

  T_mas_h = T_i + T_prima*h

  print(" Cuando t=", i/2+h,"hrs.","T =" ,T_mas_h, "°C")

  tiempo_2.append(i/2+h)

  temperatura_2.append(T_mas_h)

  T_i = T_mas_h
```

Comparación con gráfica.

```python
eje_x_2 = tiempo_1

eje_y_2 = temperatura_1

eje_x_3 = tiempo_2

eje_y_3 = temperatura_2

graficar.figure(figsize=(8,8))

graficar.plot(eje_x_2, eje_y_2, label = "Sol. Aproximada con Descenso de Temperatura Constante", linewidth = 2)

graficar.plot(eje_x_3, eje_y_3, label = "Sol. Aproximada con Descenso de Temperatura Linealmente Variable", linewidth = 2)

graficar.title (" Solución aproximada con Euler")

graficar.xlabel("Tiempo (hrs.)")

graficar.ylabel("Temperatura (°C)")

graficar.legend()

graficar.grid(True)

graficar.show()
```
